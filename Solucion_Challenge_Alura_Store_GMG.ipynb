### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# --- ANÁLISIS 1: FACTURACIÓN TOTAL POR SUCURSAL ---

# El primer paso es entender el volumen de negocio. Calculamos los ingresos totales de cada sucursal.
# Para ello, sumamos todos los valores de la columna 'Precio'.

ingresos_sucursal1 = ventas_sucursal1['Precio'].sum()
ingresos_sucursal2 = ventas_sucursal2['Precio'].sum()
ingresos_sucursal3 = ventas_sucursal3['Precio'].sum()
ingresos_sucursal4 = ventas_sucursal4['Precio'].sum()

# Presentamos los resultados de forma clara y formateada para facilitar la comparación.
print("--- Reporte de Ingresos Totales por Sucursal ---")
print(f"Sucursal 1: ${ingresos_sucursal1:,.2f}")
print(f"Sucursal 2: ${ingresos_sucursal2:,.2f}")
print(f"Sucursal 3: ${ingresos_sucursal3:,.2f}")
print(f"Sucursal 4: ${ingresos_sucursal4:,.2f}")
```

# 2. Ventas por categoría

```python
# --- ANÁLISIS 2: CATEGORÍAS DE PRODUCTOS MÁS POPULARES ---

# Contamos las ventas por categoría para cada sucursal y las ordenamos de mayor a menor.
top_categorias_s1 = ventas_sucursal1['Categoría del Producto'].value_counts()
top_categorias_s2 = ventas_sucursal2['Categoría del Producto'].value_counts()
top_categorias_s3 = ventas_sucursal3['Categoría del Producto'].value_counts()
top_categorias_s4 = ventas_sucursal4['Categoría del Producto'].value_counts()

# Mostramos las 3 categorías principales (el Top 3) para cada sucursal.
print("--- Top 3 Categorías Más Vendidas por Sucursal ---")
print("\n--- Sucursal 1 ---")
display(top_categorias_s1.head(3))
print("\n--- Sucursal 2 ---")
display(top_categorias_s2.head(3))
print("\n--- Sucursal 3 ---")
display(top_categorias_s3.head(3))
print("\n--- Sucursal 4 ---")
display(top_categorias_s4.head(3))
```

# 3. Calificación promedio de la tienda


```python
# --- ANÁLISIS 3: SATISFACCIÓN DEL CLIENTE (CALIFICACIÓN PROMEDIO) ---

# Calculamos la calificación promedio para cada sucursal usando la función .mean().
calif_promedio_s1 = ventas_sucursal1['Calificación'].mean()
calif_promedio_s2 = ventas_sucursal2['Calificación'].mean()
calif_promedio_s3 = ventas_sucursal3['Calificación'].mean()
calif_promedio_s4 = ventas_sucursal4['Calificación'].mean()

# Presentamos los promedios formateados a dos decimales para una lectura clara.
print("--- Calificación Promedio por Sucursal (escala de 1 a 5) ---")
print(f"Sucursal 1: {calif_promedio_s1:.2f} estrellas")
print(f"Sucursal 2: {calif_promedio_s2:.2f} estrellas")
print(f"Sucursal 3: {calif_promedio_s3:.2f} estrellas")
print(f"Sucursal 4: {calif_promedio_s4:.2f} estrellas")
```

# 4. Productos más y menos vendidos

```python
# --- ANÁLISIS 4: PRODUCTOS ESTRELLA Y REZAGADOS ---

# Analizamos el rendimiento a nivel de producto individual.
# Usamos .value_counts() para obtener la lista ordenada de ventas por producto.
ventas_por_producto_s1 = ventas_sucursal1['Producto'].value_counts()
ventas_por_producto_s2 = ventas_sucursal2['Producto'].value_counts()
ventas_por_producto_s3 = ventas_sucursal3['Producto'].value_counts()
ventas_por_producto_s4 = ventas_sucursal4['Producto'].value_counts()

# Usamos .index[0] para el nombre del más vendido y .iloc[0] para su conteo.
# Usamos .index[-1] para el nombre del menos vendido y .iloc[-1] para su conteo.
print("--- Reporte de Productos Más y Menos Vendidos ---")
print("\n--- Sucursal 1 ---")
print("Producto Estrella:", ventas_por_producto_s1.index[0], f"({ventas_por_producto_s1.iloc[0]} ventas)")
print("Producto Rezagado:", ventas_por_producto_s1.index[-1], f"({ventas_por_producto_s1.iloc[-1]} ventas)")

print("\n--- Sucursal 2 ---")
print("Producto Estrella:", ventas_por_producto_s2.index[0], f"({ventas_por_producto_s2.iloc[0]} ventas)")
print("Producto Rezagado:", ventas_por_producto_s2.index[-1], f"({ventas_por_producto_s2.iloc[-1]} ventas)")

print("\n--- Sucursal 3 ---")
print("Producto Estrella:", ventas_por_producto_s3.index[0], f"({ventas_por_producto_s3.iloc[0]} ventas)")
print("Producto Rezagado:", ventas_por_producto_s3.index[-1], f"({ventas_por_producto_s3.iloc[-1]} ventas)")

print("\n--- Sucursal 4 ---")
print("Producto Estrella:", ventas_por_producto_s4.index[0], f"({ventas_por_producto_s4.iloc[0]} ventas)")
print("Producto Rezagado:", ventas_por_producto_s4.index[-1], f"({ventas_por_producto_s4.iloc[-1]} ventas)")
```

# 5. Envío promedio por tienda

```python
# --- ANÁLISIS 5: EFICIENCIA LOGÍSTICA (COSTO DE ENVÍO PROMEDIO) ---

# Finalmente, evaluamos si los costos operativos podrían ser un factor diferencial.
# Calculamos el costo promedio de envío para cada sucursal con .mean().

envio_promedio_s1 = ventas_sucursal1['Costo de envío'].mean()
envio_promedio_s2 = ventas_sucursal2['Costo de envío'].mean()
envio_promedio_s3 = ventas_sucursal3['Costo de envío'].mean()
envio_promedio_s4 = ventas_sucursal4['Costo de envío'].mean()

# Presentamos los resultados formateados como moneda para una fácil comparación.
print("--- Costo de Envío Promedio por Sucursal ---")
print(f"Sucursal 1: ${envio_promedio_s1:,.2f}")
print(f"Sucursal 2: ${envio_promedio_s2:,.2f}")
print(f"Sucursal 3: ${envio_promedio_s3:,.2f}")
print(f"Sucursal 4: ${envio_promedio_s4:,.2f}")
```
